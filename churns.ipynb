```python
import requests
import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("/content/drive/MyDrive/c_churn/Telecom/Churn.csv")
ds.head()
```

```python
ds2 = ds.drop('customerID',1)
```

```python
ds2 = ds2.drop('MultipleLines',1)
ds2 = ds2.drop('OnlineBackup',1)
ds2 = ds2.drop('DeviceProtection',1)
ds2 = ds2.drop('TechSupport',1)
ds2 = ds2.drop('StreamingMovies',1)
ds2 = ds2.drop('MonthlyCharges',1)
ds2 = ds2.drop('PaymentMethod',1)
```

```python
ds2.head()
```

```python
#from sklearn.preprocessing import LabelEncoder
#le = LabelEncoder()
#le.fit(ds2)
#ds2.head()
```

```python
ds2 = pd.get_dummies(ds2,drop_first=True)
```

```python
ds2.head()
```

```python
X = ds2.drop('Churn_Yes',1)
y = ds2['Churn_Yes']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
from sklearn import metrics
prediction_test = model.predict(X_test)
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
dec = DecisionTreeClassifier()
res_dec = dec.fit(X_train,y_train)
prediction_test = dec.predict(X_test)
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
from sklearn.svm import SVC
svc = SVC()
res_svc = svc.fit(X_train,y_train)
prediction_test = svc.predict(X_test)
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
from sklearn.linear_model import SGDClassifier
sgdc = SGDClassifier()
result_sgdc = sgdc.fit(X_train,y_train)
prediction_test = sgdc.predict(X_test)
print (metrics.accuracy_score(y_test, prediction_test))
```
